```python
import requests
import os
from dotenv import load_dotenv
import time
from datetime import datetime
import json
```

```python
load_dotenv()

api_ID = os.getenv("api_ID")
api_key = os.getenv("api_key")
```

```python
def api_request(n_pages):
    req_jobs = []
    for i in range(1, n_pages + 1):
        req = requests.get(f'https://api.adzuna.com/v1/api/jobs/br/search/{i}?app_id={api_ID}&app_key={api_key}&results_per_page=20&what=dados')
        req_jobs.append(req.json())
    return req_jobs
```

```python
x = datetime.now().strftime("%d_%m_%Y__%H_%M")
os.makedirs("data/bronze", exist_ok=True)
req_jobs = api_request(1)
with open(f"data/bronze/{x}.jsonl", "w", encoding="utf-8") as f:
    for job in req_jobs:
        job_line = json.dumps(job, ensure_ascii=False) + '\n'
        f.write(job_line)
```
